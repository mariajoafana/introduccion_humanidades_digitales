# Transformaciones y agrupación

A menudo es necesario transformar los datos, crear nuevos a partir de operaciones sobre los existentes.

````python
import pandas as pd
````

Cargar los datos

````python
datos = pd.read_csv('obras.csv',sep=';')
datos
````

## Operaciones numéricas

Podemos crear nuevos campos, usando operaciones entre diferentes columnas.  En este caso multiplicamos los valores de las columnas 'ancho' y 'alto' y lo guardamos en el campo 'area':


````python
datos['area'] = datos['ancho'] * datos['alto']
datos
````

## Operaciones con strings

### Separar con split
Cuando operamos sobre columnas que contienen strings, Pandas nos ofrece varios métodos, idénticos a los métodos de un objeto ```str``` normal (algunos de ellos mencionados en la introducción a Python).

Podemos separar un string usando un 'patrón', en este caso una palabra.  El resultado es una lista con los 'pedazos' resultandes de la fragmentacón.

Podemos usar esta operación para crear nuevos campos.  Por ejemplo, al usar la opción ```expand=True``` de ```str.split()``` podemos crear una columna por cada pedazo generado en la fragmentación.

En el caso de la columna 'medium', podemos ver que las técnicas tiene una estructura común.  Están escritas en la forma "medio" sobre "soporte".  Podemos usar la palabra "sobre" como separador para extraer el "medio" y el "soporte".

````python
datos['medium']
````

````python
datos['tecnica'] = datos['medium'].str.split(' sobre ', expand=True)[0] ## Extraemos el primer elemento de la lista resultante del split
datos['soporte'] = datos['medium'].str.split(' sobre ', expand=True)[1] ## Extraemos el segundo elemento resultante del split
datos
````

Eliminar los espacios de la columna 'soporte', usando la función ```str.strip()```

````python
datos['soporte'] = datos['soporte'].str.strip()
datos
````

## Agrupación


La agrupación es uno de los procesos más comunes con datos estructurados.  En pandas tenemos la función ```GroupBy```, que es un método tanto de ```DataFrame``` como de ```Series```.

En el siguiente ejemplo agrupamos por el campo 'soporte', creado en el paso anterior.

````python
grupo = datos.groupby('soporte')
grupo
````

````python
grupo.groups
````

El objeto retornado por ```groupby``` es un objeto especial de tipo ```DataFrameGroupBy``` (cuando se aplica a un ```DataFrame```) o de tipo ```SeriesGroupBy``` cuando se aplica a una ```Series```.


En el siguiente ejemplo, usamos los índices guardados en el grupo 'lienzo' para obtener las filas correspondientes en el dataframe 'datos':


Podemomos conocer la cantidad de grupos con la función ```len()```, la misma función que usamos para saber la longitud de las listas.

````python
len(grupo)
````

Los nombres de los grupos los modemos conocer accediendo al método ```keys()``` de ```grupo.groups```:

````python
grupo.groups.keys()
````



Este objeto tiene una propiedad llamada ```groups``` que contiene los índices de cada grupo.  Es decir que los datos de las columnas del ```DataFrame``` o ```Series``` agrupao no quedan en el objeto ```GroupBy``` sino los índices de cada fila.  Cada grupo tiene un nombre, que es el valor del campo por el cual se realizó la agrupación.  En este caso son 'lienzo' y 'papel', los dos únicos valores resultantes de la agrupación.

````python
# obtenemos los índices del grupo
lienzo_indices = grupo.groups['lienzo']
# usamos los índices para obtener las filas del dataframe original
obras_lienzo = datos.loc[lienzo_indices]
# mostramos el resultado
obras_lienzo
````

Sin embargo, el objeto ```DataFrameGroupBy``` tiene métodos para obtener estas mismas filas más fácilmente, como el método ```get_group()```

````python
grupo.get_group('papel')
````

Obtener el conteo de los elementos en cada grupo.  El objeto retornado por esta función es una serie (```Series```)

````python
grupo.size().sort_values(ascending=False)
````

También podemos acceder a las columnas del dataframe original usando la sintáxis de ```[]```.  En este ejemplo calculamos el promedio de los valores en el campo 'area' en cada grupo, usando en método ```mean()```:

````python
grupo['area'].mean()
````

## Transformaciones

Pandas ofrece varios métodos para transformar datos.  Por ejemplo los métodos ```transform()``` y ```apply()``` aceptan funciones que retornan.  El resultado de la ejecución de la función se aplica a cada dato.

Primero definimos una función.  En este caso una función sencilla, que divide el área en 10000 para cambiar de centímetros cuadrados (cm2) a metros cuadrados (m2):

````python
# Definimos una función que acepta centímetros cuadrados
# y retorna metros cuadrados

def cm2_to_m2(area):
  return area / 10000
````

````python
cm2_to_m2(70000)
````

Podemos usar la función ```transform``` sobre la columna área.  El parámetro es la función (nótese que pasamos la función sin los ()).  ```transform()``` se encarga de ejecutar la función con cada valor en la columna área y retornar una nueva columna o ```Series```.

````python
datos['area'].transform(cm2_to_m2)
````

También existe la función ```apply()```, que también acepta una función como parámetro.  En este caso el resultado es el mismo que con ```transform()```.

````python
datos['area'].apply(cm2_to_m2)
````

En Python exsiten las funciones lambda, o funciones anónimas.  Son funciones que no tienen nombre, y que se ejecutan en el mismo *lugar* en el que se definen.

Las funciones normales de Python se declaran con un nombre, y se ejecutan en un *lugar* diferente al *lugar* en el cual se definen (con *lugar* nos referimos a cualquier línea o punto en el programa).

La función ```cm2_to_m2``` es una función tradicional, ya que tienen nombre y se llama en un lugar diferente al de su definición.

Una función lambda tiene esta forma:

```
lambda argumento : expresion
```
donde *argumento* es el parámetro de la función y *expresion* es la operación que se realiza sobre el argumento.  El resultado de la *expresion* es retornado por la función lambda.  Una función *lambda* se compone siempre de una sola expresión.

Por ejemplo, la función ```cm2_to_m2``` puede ser implementada como una función *lambda*:

```
lambda x: x/10000
```

Esta función puede ser pasada como un parámetro a ```transform()``` o ```apply()```:

````python
datos['area'].transform(lambda x: x/10000)
````

Lo mismo con ```apply()```:

````python
datos['area'] = datos['area'].apply(lambda x: x/10000)
````

````python
datos
````

### Iteración

Las funciones ```apply()``` y ```transform()``` son poderosas herramientas para operar sobre los datos de un ```DataFrame``` o ```Series``` con código muy conciso y funcionan para la mayoría de procesos, sin embargo, en ocaciones necesitamos iterar sobre un ```DataFrame```, ```Series``` o un ```GroupBy```.  Para esto tenemos una función que al ser llamada sobre uno de estos tipos de objetos, nos devuelve un *iterable* sobre el cual podemos realizar una iteración:

````python
for indice, fila in datos.iterrows():
  print(f'Este es el índice: {indice}')
  print(f'Este es el contenido de la fila:\n{fila}')
  print(type(fila))
````

````python
for indice, fila in datos.iterrows():
  ancho = fila['ancho']
  alto = fila['alto']
  if(ancho > alto):
    print('La orientación es horizontal')
  elif(alto > ancho):
    print('La orientación es vertical')
  else:
    print('La orientación es cuadrada')
````

Sin embargo, la documentación de Pandas advierte que el uso de este método es poco óptimo, pues la iteración es lenta, en comparación a las operaciones vistas anteriormente.

Para transformar el mismo DataFrame, creando el campo 'orientación' se puede hacer de esta forma:

Primero creamos una función que toma como parámetro la fila de un DataFrame y hace la comparación de los campos ancho y alto para saber cuál es mayor.  En cada caso del if/elif/else se retorna el valor 'vertical', 'horizontal' o 'cuadrada':

````python
def orientacion(fila):
  ancho = fila['ancho']
  alto = fila['alto']
  if ancho > alto:
    return 'horizontal'
  elif ancho < alto:
    return 'vertical'
  else:
    return 'cuadrada'
````

Luego podemos usar ```apply()``` para aplicar esta función a un dataframe, guardando el resultado en una nueva columna.

El parámetro ```axis=1``` significa que vamos a aplicar el resultado de la función a una columna, y no a una fila.

````python
datos['orientacion'] = datos.apply(orientacion, axis=1)
datos
````
