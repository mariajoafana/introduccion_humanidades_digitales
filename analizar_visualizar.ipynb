# Análisis y visualización

Pandas incluye algunas funciones de visualización.  Aunque estas son útiles para algunos tipos de gráficos, la mayoría están diseñados para datos cuantitativos.

````python
import pandas as pd
import matplotlib.pyplot as plt
````

## Cargando los datos

En este caso vamos a trabajar con un conjunto de datos con más registros.  Esta tabla es una selección de las obras de Beatriz González, provenientes del Catálogo Razonado de Beatriz González: https://badac.uniandes.edu.co

````python
obras = pd.read_csv('obras-bga.csv')
obras
````

````python
obras.dtypes
````

## Preparando los datos

Para convertir todas las columnas al tipo que necesitamos para procesar, podemos crear un diccionario con los nombres de los campos y sus tipos correspondientes y luego pasar ese diccionario a ```astype()```.

````python
tipos = {
'id':'int64',
'Files':'string',
'Title':'string',
'Description':'string',
'Creator':'string',
'Date Created':'datetime64[ns]',
'Date Created End':'datetime64[ns]',
'Contributor':'string',
'Format':'string',
'Dimensiones':'string',
'Ancho':'float64',
'Alto':'float64',
'Profundo':'float64',
'Type':'string',
'Dimensionaliad':'string',
'Is Part Of':'string',
'Is Referenced By':'string',
'Is Replaced By':'string',
'Medium':'string',
'Tecnica':'string',
'Tecnica_medio':'string',
'Tecnica_soporte':'string',
'Tecnica_objeto':'string',
'Edicion':'string'
}
````

````python
obras = obras.astype(tipos)
obras.dtypes
````

Para la primera gráfica vamos a agrupar el DataFrame por el campo 'Tecnica_medio', que contiene los medios (pigmentos, materias aplicadas sobre soportes) extraídas del campo 'Medium'.

````python
# Agrupar el dataframe por 'Tecnica_medio' y obtener el tamaño del cada grupo en una seria
grupo_medium = obras.groupby('Tecnica_medio').size()
# Ordenar descendente
grupo_medium = grupo_medium.sort_values(ascending=False)
# Seleccionar sólo los registros con valores mayores a 2
# es decir las técnicas con más de 2 obras.
grupo_medium = grupo_medium[grupo_medium > 2]
grupo_medium
````

````python
grupo_medium.to_csv('grupo_medium.csv')
````

Como el resultado es una sola serie, podemos graficarla fácilmente con un gráfico de barras:

````python
# Crear un gráfico de barras
bar = grupo_medium.plot(kind='bar')

# Con figurar las etiquetas y el título
plt.xlabel('Tecnica_medio')
plt.ylabel('Número de obras')
plt.title('Número de obras por Medium')

# Mostrar el gráfico
plt.show()
````

Con los datos de fecha podemos crear un gráfico de línea, que frecuentemente se usa para graficar series temporales.

En este caso vamos a convertir la columna 'date' a tipo fecha usando la función ```pd.to_datetime()```.  Pero también se podría usar ```astype()```

````python
# Convertir la columna 'date' a date
obras['Date Created'] = pd.to_datetime(obras['Date Created'], errors='coerce')

# Agrupar por fecha y obtener el tamaño de cada grupo.
date_group_size = obras.groupby('Date Created').size()

# Crear la visualización
date_group_size.plot()

# Etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Número de trabajos')
plt.title('Trabajos por año')

# Mostrar el gráfico
plt.show()
````

## Gráfico de dispersión (scatter plot)

Uno de los gráficos más útiles para identificar correlaciones entre dos variables es el gráfico de dispersión o scatter plot.  Si bien el conjunto de datos que tenemos es principalmente categorial, podemos hacer el experimento con dos columnas numércias.  En este caso Ancho y Alto

````python
# Podemos usar el método scatter del objeto plt
# pasamos como parámetros las columnas Ancho y Alto
plt.scatter(obras['Ancho'], obras['Alto'])

# Etiquetas
plt.xlabel('Ancho')
plt.ylabel('Alto')

# Títuloo de la gráfica
plt.title('Relación entre Ancho y Alto')
plt.show()
````

Con este gráfico podemos verificar que la tendencia de la obra de Beatriz González es a la horizoonotalidad, pues los puntos se dispersan hacia la derecha.


## Plotly

Si bien MatPlotLib es la librería más usada para visualización de datos, realmente es una librería que se usa como base para otras herramientas.

Existen varias alternativas a MatPlotLib.  En este caso vamos a usar [Plotly](https://plotly.com/), que trae ya varias herramientas fáciles de usar para crear gráficos y visualizaciones interactivas.

Plotly es realmente más que una librería.  Es un conjunto de librerías en varios lenguajes de programación.  Se puede entender mejor como un 'framework' o entorno de trabajo para visualización de datos.

Plotly express es una librería dentro de plotly, que trae ya una serie de plantillas fáciles de usar con los gráficos comunmente usados.

Para usar [Plotly Express](https://plotly.com/python-api-reference/plotly.express.html) podemos importarla de la siguiente forma

````python
# Importamos express, de la librería plotly
# De ahora en adelante se llamará px
from plotly import express as px
````

````python
fig = px.bar(grupo_medium, x=grupo_medium.index, y=grupo_medium.values)
fig.show()
````

Este gráfico es interactivo.  Puede hacer rollover sobre las barras para ver los valores asociados.

En plotly también existe el scatter plot.  En este caso podemos pasar los nombres de las columnas que contienen los valores que queremos graficar:

````python
sfig = px.scatter(obras, x='Ancho', y='Alto', hover_name='Title')
sfig.show()
````

Vamos a mapear la técnica al color de los puntos, pero primero tenemos que eliminar las filas que no tienen datos en las columnas que vamos a usar.

Al usar dropna con el parámetro subset, podemos descartarr las filas con valores NaN en cualquiera de las columnas especificadas.

````python
# Primero filtramos los datos
filtradas = obras.dropna(subset=['Ancho','Alto','Tecnica_medio','Tecnica_soporte'])
````

Ahora usamos la gráfica 'scatter' de ```plotly.exress```, solo que esta vez usamos el parámetro color, para asignarle un color a cada punto, basado en la técnica de la obra.

````python

# Creamos el gráfico de dispersion, ahora usando el parámetro  'color'
sfig = px.scatter(filtradas, x='Ancho', y='Alto', hover_name='Title', color='Tecnica_medio')

sfig.show()
````

Los gráficos de Plotly son interactivos por defecto.  Puede arrastrar el mouse para seleccionar un grupo de puntos, hacer rollover para ver los detalles de un punto.

Arriba a la derecha se despliega un menú de opciones de interacción.  Este se revela al mover el mouse sobre cualquier parte del gráfico.

### Escala de los ejes

El en siguiente ejemplo establecemos que los ejes del gráfico sean proporcionales, para entender mejor la relación entre las dimensiones.

````python

# Creamos el gráfico de dispersion, ahora usando el parámetro  'color'
sfig = px.scatter(filtradas, x='Ancho', y='Alto', hover_name='Title', color='Tecnica_medio')

sfig.update_xaxes(
    range=[1,7000],  # sets the range of xaxis
    constrain="domain",  # meanwhile compresses the xaxis by decreasing its "domain"
)
sfig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1
)
sfig.show()
````

### Categorías paralelas

````python
paralelas = px.parallel_categories(filtradas, dimensions=['Dimensionaliad', 'Tecnica_medio'], height=1200)
paralelas.show()
````

## Treemap

Esta es una visualización para conjuntos de datos jeráquicos, es decir que tenga categorías y subcategoorías.  En este caso vamos a usar algunas columnas que se puede leer como jerárquicas: 'Dimensionalidad' (si es bidimensional o tridimensionoal), luego 'Tecnica_soporte' y por último 'Tecnica_medio'.

````python
tree = px.treemap(filtradas, path=['Dimensionaliad', 'Tecnica_soporte','Tecnica_medio'])
tree.show()
````

En esta visualización podemos ver la distribución de las obras según su dimensionalidad, luego su soporte y luego sus medios.

Podemos conluir que, según los datos, la obra de Beatriz González es predominantemente bidimensional.
